```python
import serial
import pandas as pd
import time
ser = serial.Serial('com3', 9600)
counter = 0
f = open('log4.txt', 'w')
while True:
    try:
        data = ser.readline()
    except:
        break
    data = data.decode()
    print(data)
    f.write(data)
    f.write(str(time.localtime(time.time()).tm_sec)+ '\n')
f.close()
```
